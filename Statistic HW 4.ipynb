```python
Задача 1 - З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?
```

Використовуємо формулу умовної ймовірності 

P(A | B)=P(A∩B) / P(B)

де-  A - карта -туз,  B - карта -  пікова

Ймовірність того що карта пікова ( у колоді 13 пікових карт) 

P(B) = 13/ 52 = 1/4

Ймовірність того що витягнута карта туз піковий

P(A∩B) = 1/ 52

Тоді

P(A | B) = (1/52 )/ (1/4) =  4/52 = 1/13

Задача 2 - У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?

P(Дівчина | Окуляри)= P(Дівчина∩Окуляри) / P(Окуляри)

P(Дівчина∩Окуляри)= 5 / 30

P(Окуляри) = (7 + 5 )/ 30 = 12/30 = 2/5 

P(Дівчина | Окуляри) = (5/30) / (2/5 ) = 25/ 60 = 5/12
