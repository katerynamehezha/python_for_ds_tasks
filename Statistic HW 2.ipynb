**Загальна кількість куль:**

n(S)=15+5+10=30n(S) = 15 + 5 + 10 = 30

**Ймовірність витягнення білої кулі:**

P(біла)  = n(біла)/ n(S) = 15 / 30 = 1/2

**Ймовірність витягнення червоної кулі:**

P(червона)  = n(червона)/ n(S) = 5 / 30 = 1/6

**Ймовірність витягнення чорної кулі:**

P(чорна)  = n(чорна)/ n(S) = 10 / 30 = 1/3

n(A) = n(S) - кількість дефектних = 30 - 5 = 25

**P(без дефекту)** = n(A) / n(S) = 25 / 30 = 5/6 = 0.8333 (83.33%)

Задача 4 -  Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
    
    а) дорівнюватиме семи;
    б) буде не меншим за 20;
    в) буде парним.

Загальна кількість результатів кидка 2х кубиків: 

n(S)=6 * 6=36 

a)  визначити, які комбінації чисел на кубиках дають добуток 7. = це 1 * 7 чи 7 * 1.

Але але немає "7" на кубику тобто немає варіантів, що добуток буде 7

P(добуток=7)= 0 / 36 = 0

б) Визначаємо яка комбінація чисел дає добуток який  буде не меншим за 20:

4×5=20, 5×4=20, 4×6=24, 6×4=24, 5×5=25, 5×6=30, 6×5=30, 6×6=36 -   8  варіантів

P(добуток≥20)= 8 / 36 = 2/9

в) Добуток двох чисел буде парним, якщо хоча б одне з цих чисел парне.

Парні числа на кубику - 2, 4, 6

1 кубик парне число (3 варіанти), тоді на 2 кубику може випасти будь-яке число (6 варіантів):

3×6=18

1 кубик непарне число (3 варіанти), тоді на другому кубику має випасти парне число (3 варіанти):

3×3=9

Загальна кількість сприятливих випадків:

18+9=27

P(парний добуток)= 27 / 36 = 3/4
