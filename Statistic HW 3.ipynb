Англійський алфовіт має 26 літери 

1 літеру - можемо вибрати з 26 літер 

2 літеру. - з 25 

3 літеру - з 24

Кількість паролів = 26 * 25 * 24 = 15600

Задача 2 -  Є група з 8 людей. Скількома способами можна обрати комітет із 3 осіб?

Будемо використовувати формулу для обчислення комбінацій

С(n, r) = n! /  ((n -r)*r! ) 

C(8, 3) = 8! / ((8 - 3) *3!) = 8! / 3! * 5! = (8*7*6*5!) / 3! * 5! = (8*7*6 )/ 6 = 8 * 7 = 56

Задача 3 - Скількома різними способами можна розташувати на полиці 5 різних книг?

Будемо використовувати формулу факторіалу - n!

5! = 5 * 4 * 3* 2 * 1 = 120 різними способами

Задача 4 - В урні знаходяться 4 червоні, 3 сині і 2 зелені кулі. Скількома способами можна витягти 3 кулі, якщо вони повинні бути різних кольорів?

Спочатку розрахуємо кількість способів вибрати кожну кулю конкретного кольору:

Червону - C(4,1) = 4

Синю - C(3,1) = 3

Зелену - C(2,1) = 2

Загальна кількість способів - 4*3*2=24
