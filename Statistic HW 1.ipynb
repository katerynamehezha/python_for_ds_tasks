Кількість груп користувачів котрі підписались на додаток:

- **Травень 2019:** 100 користувачів
- **Червень 2019:** 150 користувачів
- **Липень 2019:** 170 користувачів

Щомісячна вартість підписки: $10 

Щомісячний рівень відтоку: 5% (залишається 95% користувачів)

**Знайти:**

Загальний дохід до травня 2021 року включно( за 24 місяці для травня 2019, 23 місяці для червня 2019 та 22 місяці для липня 2019)

- У червні 2019 року залишилось: 100×0.95=95
- У липні 2019 року залишилось: 95×0.95=90.25
- Далі продовжимо цей процес на 24 місяці.

для групи що  приєдналася в червні 2019:

- У липні 2019 року залишилось: 150×0.95=142.5
- У серпні 2019 року залишилось: 142.5×0.95=135.375
- Продовжимо цей процес на 23 місяці.

для групи що  приєдналася в червні 2019:

- У серпні 2019 року залишилось: 170×0.95=161.5
- У вересні 2019 року залишилось: 161.5×0.95=153.425
- Продовжимо цей процес на 22 місяці.

Загальний дохід: 

- Дохід від групи з травня 2019: **$22,668.91**
- Дохід від групи з червня 2019: **$22,480.34**
- Дохід від групи з липня 2019: **$12,790.12**

Загальний дохід за весь період з трьох груп: **$57,939.37**

Задача 2. - Ви — менеджер по роботі з продуктом в компанії, яка розробляє мобільні ігри. Вашою метою є збільшення кількості активних користувачів. У січні 2020 року у вас було 2000 активних користувачів. З лютого кількість активних користувачів щомісяця зростає на 3%. У той же час, 2% користувачів щомісяця перестають користуватися додатком. Якою буде кількість активних користувачів за весь період від січня 2020 року до кінця грудня 2021 року?

Щоб зробити обчислення будемо враховувати зростання(кожного місяця на 3%) і відтік( кожного місяця 2%) користувачів

**Розрахуємо чисте зростання =** (1+0.03)×(1−0.02)=1.03×0.98 = 1.0094 ( кожного місяця кількість активних користувачів зростає приблизно на 0.94%)

**Вхідні данні:**

Початкова кількість користувачів = 2000

Кількість місяців = 24

**Рішення:**

Кількість активних користувачів через t місяців: 

Користувачі(t)=2000×(1.0094)^t 

Користувачі(24)=2000×(1.0094)^24 = 2504

На кінець грудня 2021 року кількість активних користувачів буде приблизно **2504**

**Травень 2021:** Зростання на 10%

**Червень 2021:** Зростання на 10%:

1089×(1+0.10)=1089×1.10≈1197.9 =1198
